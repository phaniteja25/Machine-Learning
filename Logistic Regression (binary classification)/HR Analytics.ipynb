### Following attributes will be considered

1.Satisfaction Level
2. Average Monthly Hours
3. promotion last 5 years
4. salary

```python
df_new = df[['satisfaction_level','average_montly_hours','promotion_last_5years','salary']]
df_new.head()
```

```python
salary_dummies = pd.get_dummies(df_new.salary).astype(int)
```

```python
salary_dummies
```

```python
df_with_dummies = pd.concat([df_new,salary_dummies],axis='columns')
```

```python
df_with_dummies.head()
```

```python
df_with_dummies.drop(columns="salary",axis=1,inplace=True)
```

```python
df_with_dummies.head()
```

```python
X = df_with_dummies
X.head()
```

```python
y = df.left
y.head()
```

### Prediction 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test,y_test)
```
